```python
import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Text
text = "Amazon is planning to open a new headquarters in Arlington, Virginia by 2025, creating over 25,000 jobs and investing $2.5 billion."

# Process text
doc = nlp(text)

# Print entities
print("spaCy NER:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

# Download the necessary NLTK data
nltk.download("punkt")
nltk.download("maxent_ne_chunker")
nltk.download("words")
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab')
# Download the missing 'maxent_ne_chunker_tab' resource
nltk.download('maxent_ne_chunker_tab') # This line was added to fix the LookupError

# Sample sentence
text = "Amazon is planning to open a new headquarters in Arlington, Virginia by 2025, creating over 25,000 jobs and investing $2.5 billion."

# Tokenization and Named Entity Recognition
tokens = word_tokenize(text) # [cite: 7]
tags = pos_tag(tokens) # [cite: 7]
chunked = ne_chunk(tags) # [cite: 7]

print(chunked)
# chunked.draw() # This opens a tree structure visualization [cite: 7]
```

```python
from transformers import pipeline

ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", grouped_entities=True)

print("BERT NER:")
entities = ner_pipeline(text)
for ent in entities:
    print(ent)
```
